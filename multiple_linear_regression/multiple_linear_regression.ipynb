```python
class MyMultipleLinearRegression:
    def __init__ (self):
        self.weights=None
        self.bias=None

    def fit (self, x_input,y_ouput):
        x_input= np.insert(x_input,0,1,axis=1)
        
        self.coef=np.dot(np.dot(np.linalg.inv(np.dot(x_input.T,x_input)),x_input.T),y_ouput)
        self.bias=self.coef[0]
        self.weights=self.coef[1:]

    def predict(self,x_train):
        return np.dot(np.insert(x_train,0,1,axis=1),self.coef)
```

```python
lr1=MyMultipleLinearRegression()
```

```python
from sklearn.datasets import load_diabetes
x,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr1.fit(x_train,y_train)
```

```python
print(lr1.weights)
```

```python
lr1.predict(x_test)
```
